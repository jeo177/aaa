## Grounding DINO Demo

```python
%cd {HOME}

from mmdet.apis import DetInferencer
from PIL import Image

config_file = 'mmdetection/configs/grounding_dino/grounding_dino_swin-b_pretrain_mixeddata.py'
checkpoint_file = 'weights/groundingdino_swinb_cogcoor_mmdet-55949c9c.pth'
inferencer = DetInferencer(config_file, checkpoint_file)
```

```python
img = 'data/dog-3.jpeg'
result = inferencer(img, out_dir = './output', texts='chair')
Image.open('./output/vis/dog-3.jpeg')
```

```python
img = 'data/dog-3.jpeg'
result = inferencer(img, out_dir = './output', texts='chair with man sitting on it')
Image.open('./output/vis/dog-3.jpeg')
```
